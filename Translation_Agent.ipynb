```python
import pdfplumber

pdf_file_path = '/content/drive/My Drive/Translation/English_Source.pdf'  # Replace with your PDF file name

extracted_text = ""
with pdfplumber.open(pdf_file_path) as pdf:
    for page in pdf.pages:
        extracted_text += page.extract_text()

print("Total Characters Extracted:", len(extracted_text))
print("Preview of Extracted Text:", extracted_text[:500])  # Show the first 500 characters
```

```python
chunk_size = 1000  # Adjust this size based on API limits
text_chunks = [extracted_text[i:i+chunk_size] for i in range(0, len(extracted_text), chunk_size)]

print(f"Total Chunks: {len(text_chunks)}")
print("Preview of First Chunk:", text_chunks[0])
```

```python
import src.translation_agent as ta  # Ensure your translation agent is properly imported

source_lang = "English"
target_lang = "Spanish"
country = "Spain"  # Optional

translated_text = ""
for chunk in text_chunks:
    translation = ta.translate(source_lang, target_lang, chunk, country)
    translated_text += translation + "\n"

print("Translation Preview:", translated_text[:500])  # Display first 500 characters of the translation
```

```python
from docx import Document

# Load the text file
text_file_path = "translated_text.txt"  # Path to your text file
word_file_path = "translated_text.docx"  # Desired output Word file

# Create a Word document
doc = Document()

# Read the text file and add its content to the Word document
with open(text_file_path, "r", encoding="utf-8") as file:
    for line in file:
        doc.add_paragraph(line.strip())

# Save the Word document
doc.save(word_file_path)
print(f"Word document saved as '{word_file_path}'")
```
